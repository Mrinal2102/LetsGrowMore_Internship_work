# Data Preprocessing - To convert Categorical Variable

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
```

```python
LE = LabelEncoder()
```

```python
df.iloc[:,-1] = LE.fit_transform(df.iloc[:,-1])
```

```python
df
```

# Do Train Test Split

```python
X = df.iloc[:,:4] # Independent Variable
X.head()
```

```python
y = df.iloc[:,-1] # dependent variable
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X.shape
```

```python
y_train.head()
```

```python
y.shape
```

# Model Building

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

# Compare actual and predicted result

```python
y_pred = dt.predict(X_test)
y_pred
```

```python
y_test = np.array(y_test)
y_test
```

# Accuracy of the Decision Tree Model

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred, y_test)
```

# Confusion Matrix for Decision Tree Model

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_pred, y_test)
```

Confusion matrix shows the ways in which our classification model is confused when it makes predictions

# Classification Report of Decision Tree Model

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_pred, y_test))
```
